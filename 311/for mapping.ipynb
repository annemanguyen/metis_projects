```python
top_n = 5
complaints_dict = {n: percentage.T[col].nlargest(top_n)
                  for n, col in enumerate(percentage.T)}
```

```python
first = []
second = []
third = []
fourth = []
fifth = []

firstpercent = []
secondpercent = []
thirdpercent = []
fourthpercent = []
fifthpercent = []

firstcount = []
secondcount = []
thirdcount = []
fourthcount = []
fifthcount = []
```

```python
for i in range(len(percentage.T.columns)):
    first.append(percentage.T[i].nlargest(5).index[0])
    firstpercent.append(percentage.T[i].nlargest(5)[0])
    firstcount.append(complaints.T[i].nlargest(5)[0])
    
    second.append(percentage.T[i].nlargest(5).index[1])
    secondpercent.append(percentage.T[i].nlargest(5)[1])
    secondcount.append(complaints.T[i].nlargest(5)[1])
    
    third.append(percentage.T[i].nlargest(5).index[2])
    thirdpercent.append(percentage.T[i].nlargest(5)[2])
    thirdcount.append(complaints.T[i].nlargest(5)[2])
    
    fourth.append(percentage.T[i].nlargest(5).index[3])
    fourthpercent.append(percentage.T[i].nlargest(5)[3])
    fourthcount.append(complaints.T[i].nlargest(5)[3])
    
    fifth.append(percentage.T[i].nlargest(5).index[4])
    fifthpercent.append(percentage.T[i].nlargest(5)[4])
    fifthcount.append(complaints.T[i].nlargest(5)[4])
```

```python
fifthcount[0:5]
```

```python
for_map = pd.DataFrame()
```

```python
for_map['ntacode'] = nb.ntacode
for_map['ntaname'] = acs.GeogName
for_map['first'] = first
for_map['firstcount'] = firstcount
for_map['firstpercent'] = firstpercent
for_map['second'] = second
for_map['secondcount'] = secondcount
for_map['secondpercent'] = secondpercent
for_map['third'] = third
for_map['thirdcount'] = thirdcount
for_map['thirdpercent'] = thirdpercent
for_map['fourth'] = fourth
for_map['fourthcount'] = fourthcount
for_map['fourthpercent'] = fourthpercent
for_map['fifth'] = fifth
for_map['fifthcount'] = fifthcount
for_map['fifthpercent'] = fifthpercent
```

```python
counts = ['firstcount', 'secondcount', 'thirdcount', 'fourthcount', 'fifthcount']
```

```python
for_map[counts] = for_map[counts].astype(int)
```

```python
percents = ['firstpercent', 'secondpercent', 'thirdpercent', 'fourthpercent', 'fifthpercent']
```

```python
for_map[percents] = for_map[percents]*100
```

```python
for_map[percents] = for_map[percents].round(1)
```

```python
for_map['rate']  = nb.rate*100
```

```python
for_map['rate'] = for_map.rate.round(1)
```

```python
for_map.head()
```

```python
for_map.to_csv('for_map.csv')
```
