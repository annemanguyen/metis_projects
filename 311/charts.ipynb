```python
import pandas as pd
from collections import Counter
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
c = pd.read_pickle('full.pkl')
```

```python
categories = Counter(c.category.values)
```

```python
cat = pd.DataFrame.from_dict(categories, orient='index').reset_index()
```

```python
cat.columns = ['categories', 'count']
```

```python
cat.sort_values(by='count', inplace=True, ascending=False)
```
